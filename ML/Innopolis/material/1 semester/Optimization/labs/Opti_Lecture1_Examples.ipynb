# Example 1 : resources allocation

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(-1, 10, 0.01)
x2 = np.arange(-1, 12, 0.01)

plt.rcParams.update({'font.size': 25, 'lines.linewidth': 2})
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.xticks(np.arange(0, 8, 1))
plt.yticks(np.arange(0, 6, 1))
plt.axis([-0.5, 7, -0.5, 5])

plt.plot(x1, (13-2*x1)/3, 'b')
plt.plot(x1, (11-2*x1), 'k')
plt.plot(x1, (9-x1)/3, 'y')
plt.plot(x1, np.zeros(len(x1)), 'g')
plt.plot(np.zeros(len(x2)), x2, 'm')

plt.plot(x1, (1200-300*x1)/400, 'r-.')
plt.plot(x1, (1900-300*x1)/400, 'r--')
plt.legend(['$2x_1+3x_2 = 13$', '$2x_1 + x_2 = 11$', '$x_1+3x_2 = 11$', '$x_2=0$', '$x_1 = 0$', '$300x_1+400x_2 = 1200$', '$300x_1+400x_2 = 1900$'], loc='upper right', prop={'size': 10})

plt.grid()
```

## Solve it using cvxpy
 - (CVXPY is an open source Python-embedded modeling language for convex optimization problems.)
 - Link: https://www.cvxpy.org/

```python
# Import packages.
import cvxpy as cp
import numpy as np
```

```python
# Let us build our Problem of Resources allocation at hand
m = 5                              # number of inequality constraints
n = 2                              # dimension of the problem
c = np.array([300, 400])           # vector of coefficients of the objective function
A = np.array([[2,3],[2,1],[1,3],[-1,0],[0, -1]]) # coeff matrix of constraints
b = np.array([13,11,9,0,0])                      # vector of independent terms
```

```python
# Define and solve the CVXPY problem.
x = cp.Variable(n)
prob = cp.Problem(cp.Maximize(c.T@x),
                 [A @ x <= b])
prob.solve()

# Print result.
print("\nThe optimal value is", prob.value)
print("A solution x is")
print(x.value)
# print("A dual solution is")
# print(prob.constraints[0].dual_value)
```

```python
# Let us build our Problem of Resources allocation at hand
                  # vector of independent terms
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(-1, 10, 0.01)
x2 = np.arange(-1, 11, 0.01)

fig, ax = plt.subplots()
plt.rcParams.update({'font.size': 25, 'lines.linewidth': 2})


ax.plot(x1, 10-2*x1, 'b')
ax.plot(x1, (8-x1)/2, 'k')
ax.plot(x1, 7-x1, 'y')
ax.plot(x1, np.zeros(len(x1)), 'g')
ax.plot(np.zeros(len(x2)), x2, 'm')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

plt.xticks(np.arange(0, 10, 1))
plt.yticks(np.arange(0, 11, 1))
plt.axis([-0.5, 10, -0.5, 11])


ax.plot(x1, (12-4*x1)/3, 'r-.')

plt.legend(['$2x_1+x_2 = 10$', '$x_1 + 2x_2 = 8$', '$x_1+x_2 = 7$', '$x_2=0$', '$x_1 = 0$', '$4x_1+3x_2 = 12$'], loc='upper right', prop={'size': 10})
plt.grid()
```

```python
# Let us build our Problem of Resources allocation at hand
m = 5                              # number of inequality constraints
n = 2                              # dimension of the problem
c = np.array([4, 3])           # vector of coefficients of the objective function
A = np.array([[2,1],[1,2],[1,1],[1,0],[0, 1]]) # coeff matrix of constraints
b = np.array([10,8,7,0,0])                      # vector of independent terms
```

```python
# Define and solve the CVXPY problem.
x = cp.Variable(n)
prob = cp.Problem(cp.Minimize(c.T@x),
                 [A @ x >= b])
prob.solve()

# Print result.
print("\nThe optimal value is", prob.value)
print("A solution x is")
print(x.value)
# print("A dual solution is")
# print(prob.constraints[0].dual_value)
```
