```python
import numpy as np
import pandas as pd

pd.set_option("display.precision", 2)
pd.set_option('display.max_columns', 100)
```

```python
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/"
```

```python
df = pd.read_csv(DATA_URL + "telecom_churn.csv")
df.head()
```

|Feature Name|Description|Value Type|Statisrical Type|
|------------|-----------|----------|----------------|
|State| Штат США (аббревиатура)|string|categorical|
|Account length|Длина аккаунта (как долго клиент сотрудничает с этой компанией)|Numerical|quantitative|
|Area code|Код города (префикс номера телефона)|Numerical|categorical|
|International plan|Международный план|string:yes/no|categorical/boolen|
|Voice mail plan|План голосовой почты|string:yes/no|categorical/boolen|
|Number vmail messages|Количество сообщений vmail|Numerical|quantitative|
|Total day minutes|Всего минут за день|Numerical|quantitative|
|Total day calls|Всего звонков за день|Numerical|quantitative|
|Total day charge|Общая стоимость дня|Numerical|quantitative|
|Total eve minutes|Всего минут вечером|Numerical|quantitative|
|Total eve calls|Всего звонков вечером|Numerical|quantitative|
|Total eve charge|Общая стоимость вечера|Numerical|quantitative|
|Total night minutes|Всего ночных минут|Numerical|quantitative|
|Total night calls|Всего ночных звонков|Numerical|quantitative|
|Total night charge|Общая стоимость ночи|Numerical|quantitative|
|Total intl minutes|Всего международных минут|Numerical|quantitative|
|Total intl calls|Всего международных звонков|Numerical|quantitative|
|Total intl charge|Общие расходы на международные звонки|Numerical|quantitative|
|Customer service calls|Звонки в службу поддержки клиентов|Numerical|categorical/ordinal|
|Churn (target)| отток (True - клиент ушёл, False - лояльный клиент)|boolean|boolean|

```python
df["Churn"] = df["Churn"].astype("int64")
```

```python
df.describe(include=['object','bool'])
```

Мы видим, что соотношение целевого признако неравномерно (целевые классы неравномерны)

```python
df.sort_values(by=['Churn','Total day charge'], ascending=[True, False])
```

```python
df[df['Churn']==1].describe(include=np.number)
```

```python
df[(df['Churn']==0) & (df['International plan'] == 'No')].describe()
```

```python
df[df['State'].apply(lambda state: state[0]=='W')]
```

```python
df[['International plan', 'Voice mail plan']] = df[['International plan', 'Voice mail plan']].map(lambda x: False if x=='No' else True)
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].agg(['mean','std','min','max'])
```

```python
pd.crosstab(df['Churn'], df['International plan'], normalize=True)
```

```python
pd.crosstab(df['Churn'], df['Voice mail plan'], normalize=True)
```

```python
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'], ['Area code'], aggfunc=['mean'])
```

```python
df['Total calls'] = df['Total day calls'] + df['Total eve calls'] + df['Total night calls']
df['Total charge'] = df['Total day charge'] + df['Total eve charge'] + df['Total night charge']
```
